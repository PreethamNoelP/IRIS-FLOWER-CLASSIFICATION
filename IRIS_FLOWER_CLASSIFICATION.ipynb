*Data Visualization*

```python
def histogram(column_name):
  sns.histplot(data_set=iris[column_name],bins=20,kde=True)
  plt.title(column_name+' Distribution')
  plt.show()
```

```python
def count_plot(column):
  plt.figure(figsize=(10, 6))
  sns.countplot(x=column, data=iris)
  plt.title(f"Count Plot - {column}")
  plt.xlabel(column)
  plt.ylabel("Count")
  plt.xticks(rotation=90)
  plt.show()
```

```python
iris['Species'].value_counts().reset_index()
```

```python
count_plot('Species')
```

```python
count_plot('PetalLengthCm')
```

```python
def generate_histogram(iris, SepalLengthCm, SepalWidthCm, bins=10):
    plt.figure(figsize=(10, 6))
    plt.hist(iris[SepalLengthCm], bins=bins, alpha=0.5, label=SepalLengthCm)
    plt.hist(iris[SepalWidthCm], bins=bins, alpha=0.5, label=SepalWidthCm)
    plt.xlabel("Sepal Length (cm)")
    plt.ylabel("Sepal Width (cm)")
    plt.title(f"Histogram of {SepalLengthCm} and {SepalWidthCm}")
    plt.legend()
    plt.show()
generate_histogram(iris, 'SepalLengthCm', 'SepalWidthCm')
```

```python
def generate_histogram(iris, PetalLengthCm, PetalWidthCm, bins=10):
    plt.figure(figsize=(10, 6))
    plt.hist(iris[PetalLengthCm], bins=bins, alpha=0.5, label=PetalLengthCm)
    plt.hist(iris[PetalWidthCm], bins=bins, alpha=0.5, label=PetalWidthCm)
    plt.xlabel("Petal Length (cm)")
    plt.ylabel("Petal Width (cm)")
    plt.title(f"Histogram of {PetalLengthCm} and {PetalWidthCm}")
    plt.legend()
    plt.show()
generate_histogram(iris, 'PetalLengthCm', 'PetalWidthCm')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Species", y="SepalLengthCm", data=iris)
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="SepalLengthCm", y="SepalWidthCm", hue="Species", data=iris)
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.show()
```

*Dividing the data in order to train and test the dataset*

```python
model_svc.fit(X_train, Y_train)
```
